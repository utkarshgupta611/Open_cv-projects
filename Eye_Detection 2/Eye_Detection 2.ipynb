## IMPORT LIBRARY

```python
import cv2
import matplotlib.pyplot as plt
```

## READ DATA

```python
image = cv2.imread('people1.jpg')
```

```python
image.shape
```

```python
plt.imshow(image)
plt.show()
```

## GRAY SCALE


we need to scale the image to gray to decrease the demension

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray)
```

```python
image_gray.shape
```

## Detecting Face and eye using haarcascade classifier

```python
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_detector = cv2.CascadeClassifier('haarcascade_eye.xml') 
```

```python
face_detection = face_detector.detectMultiScale(image_gray, scaleFactor = 1.2)
eye_detection = eye_detector.detectMultiScale(image_gray, scaleFactor = 1.2)
```

```python
len(eye_detection)
```

```python
len(face_detection)
```

```python
for (x, y, w, h) in face_detection:
  # print(x, y, w, h)
  cv2.rectangle(image, (x, y), (x + w, y + h), (255, 2, 0), 4) 
plt.imshow(image)
```

```python
for (x, y, w, h) in eye_detection:
  # print(x, y, w, h)
  cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 250), 4) 
plt.imshow(image)
```
