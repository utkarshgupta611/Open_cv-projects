## IMPORT LIBRARY

```python
import cv2
import matplotlib.pyplot as plt
```

## READ DATA

```python
image = cv2.imread('people1.jpg')
```

```python
image.shape
```

```python
plt.imshow(image)
plt.show()
```

```python
image = cv2.resize(image, (800, 600))
```

```python
image.shape
```

```python
plt.imshow(image)
plt.show()
```

## GRAY SCALE

we need to scale the image to gray to decrease the demension

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_gray)
```

```python
image_gray.shape
```

## Detecting Face using haarcascade classifier

```python
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
detection = face_detector.detectMultiScale(image_gray, scaleFactor = 1.09)
```

```python
detection
```

```python
len(detection)
```

## FACE DETECTION

```python
for (x, y, w, h) in detection:
  # print(x, y, w, h)
  cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 4) 
plt.imshow(image)
```

## lets try with other example

```python
image2 = cv2.imread('people2.jpg')
image_gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
detection = face_detector.detectMultiScale(image_gray2, scaleFactor = 1.09, minNeighbors=7,
                                           minSize=(20,20), maxSize=(100, 100))
for (x, y, w, h) in detection:
  # print(x, y, w, h)
  cv2.rectangle(image2, (x, y), (x + w, y + h), (0, 255, 0), 4) 
plt.imshow(image2)
```
